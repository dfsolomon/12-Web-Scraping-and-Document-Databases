```python
#mars hemispheres
#get titles of images
home = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
response = requests.get(home)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
#mars hemispheres  

#empry arrays to hold names and urls
picturename = []
pictureurl = []
#set up splinter
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

browser.visit(home)
#lots of sleep functions to allow for load times 
time.sleep(3)
browser.click_link_by_partial_text('Cerberus') 
response = requests.get(browser.url)
time.sleep(3)
#soup to get titles of pictures
soup = BeautifulSoup(response.text, 'html.parser')
title = soup.find('h2', class_= "title").text
picturename.append(title)
time.sleep(3)
#open picture and take URL and append to array
browser.click_link_by_partial_text('Sample')
pictureurl.append(browser.url)
time.sleep(3)

#repeat

browser.visit(home)
time.sleep(3)
browser.click_link_by_partial_text('Schiaparelli')
response = requests.get(browser.url)
time.sleep(3)
soup = BeautifulSoup(response.text, 'html.parser')
title = soup.find('h2', class_= "title").text
picturename.append(title)
time.sleep(3)
browser.click_link_by_partial_text('Sample')
pictureurl.append(browser.url)
time.sleep(3)

browser.visit(home)
time.sleep(3)
browser.click_link_by_partial_text('Syrtis')
response = requests.get(browser.url)
time.sleep(3)
soup = BeautifulSoup(response.text, 'html.parser')
title = soup.find('h2', class_= "title").text
picturename.append(title)
time.sleep(3)
browser.click_link_by_partial_text('Sample')
pictureurl.append(browser.url)
time.sleep(3)

browser.visit(home)
time.sleep(3)
browser.click_link_by_partial_text('Marineris')
response = requests.get(browser.url)
time.sleep(3)
soup = BeautifulSoup(response.text, 'html.parser')
title = soup.find('h2', class_= "title").text
picturename.append(title)
time.sleep(3)
browser.click_link_by_partial_text('Sample')
pictureurl.append(browser.url)
time.sleep(3)
browser.quit()

#findings
results = {
    "Picture name": picturename,
    "Picture URL": pictureurl
}
print(results)
```
